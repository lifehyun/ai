* inner join이란 ?
   - 아래와 같은 표를 가정하에 a와 b 테이블에 동일한 데이터가 들어갈시 그것들만 join(합쳐준다)라는 뜻

 * a라는 데이터 테이블   
  - (1 a)
  - (2 b)
  - (3 c)

 * b라는 데이터 테이블
  - (1 a)
  - (2 b)
  - (4 e)

* full outer join 외부조인
 - left 왼쪽만 조인해주는 명령
 - right 오른쪽만 조인해주는 명령


```python
from datetime import datetime
import pandas as pd
import numpy as np

# 1.  데이터터프레임 위/아래 붙이기 (concat)

# df1, df2 데이터프레임을 위/아래 방향으로 연결하세요. (위에 df1이 있고 그 아래에 df2가 있어야합니다.)

columns = ['open','high', 'low','close']

index1 = [
datetime(2022, 9, 8),
datetime(2022, 9, 7),
datetime(2022, 9, 6),
datetime(2022, 9, 5),
datetime(2022, 9, 2)
]

index2 = [
datetime(2022, 9, 1),
datetime(2022, 8, 31),
datetime(2022, 8, 30),
datetime(2022, 8, 29),
datetime(2022, 8, 26)
]

df1 = pd.DataFrame(data=np.arange(0, 20).reshape(5,4), index=index1, columns=columns)
df2 = pd.DataFrame(data=np.arange(20,40).reshape(5,4), index=index2, columns=columns)

df1
df2
df1_1 = pd.concat([df1,df2], axis=0)
print("1번",df1_1)
# 2.  데이터터프레임 위/아래 붙이기 (concat)
# 위 df1, df2 데이터프레임을 위/아래 방향으로 연결하세요. (위에 df2가 있고 그 아래에 df1가 있어야합니다.)

df2_1 = pd.concat([df1,df2], axis=0)
print("2번",df2_1)

# 3.  데이터프레임 좌/우로 붙이기 (concat)
# 다음 df1, df2 데이터프레임을 좌/우로 붙여 새로운 데이터프레임을 생성하세요.

index = [
datetime(2022, 9, 8),
datetime(2022, 9, 7),
datetime(2022, 9, 6),
datetime(2022, 9, 5),
datetime(2022, 9, 2)
]

df1 = pd.DataFrame(data=np.arange(0, 10).reshape(5,2), index=index, columns=['open', 'high'])
df2 = pd.DataFrame(data=np.arange(10,20).reshape(5,2), index=index, columns=['low', 'close'])
df3_1 = pd.concat([df1,df2], axis=1)
print("3번",df3_1)


# 4. DataFrame.concat
# 다음 df1, df2 데이터프레임을 위/아래로 연결하세요. concat 메서드에서 join='outer'를 사용하세요.

index1 = [
     datetime(2022, 9, 8),
     datetime(2022, 9, 7)
]

index2 = [
     datetime(2022, 9, 6),
     datetime(2022, 9, 5)
]
df1 = pd.DataFrame(np.arange(0, 6).reshape(2,3), index=index1, columns=['open', 'high', 'low'])
df2 = pd.DataFrame(np.arange(6,12).reshape(2,3), index=index2, columns=['high', 'low', 'close'])

df4 = pd.concat([df1,df2],join='outer')
print("4번",df4)
# 5.  DataFrame.concat (join parameter)

# 위 df1, df2 데이터프레임을 위/아래로 연결하세요. concat 메서드에서 join='inner'를사용하세요.
df1 = pd.DataFrame(np.arange(0, 6).reshape(2,3), index=index1, columns=['open', 'high', 'low'])
df2 = pd.DataFrame(np.arange(6,12).reshape(2,3), index=index2, columns=['high', 'low', 'close'])

# 6. pd.concat series

# 다음 s1, s2 시리즈 객체를 위/아래로 연결하세요.

s1 = pd.Series(['a', 'b'])
s2 = pd.Series(['c', 'd'])

df6 = pd.concat([s1,s2])
print("6번",df6)
# 6번 문제 자료로 다음 s1, s2 시리즈 객체를 좌/우로 연결하세요.
pd.concat([s1,s2],axis = 0)

# 7. pd.concat series
df7 = pd.concat([s1,s2], axis=1)

print(f'7번{df7}')

# 8. pd.concat & ignore_index
# 다음 df1, df2를 concat 메서드로 위/아래로 붙일 때 인덱스를 무시하고 연결하세요.

df1 = pd.DataFrame(np.arange(0, 4).reshape(2,2))
df2 = pd.DataFrame(np.arange(4, 8).reshape(2,2))

df8 = pd.concat([df1,df2],ignore_index= 1  )
print(df8)

# 9.  DataFrame.groupby mean
# 다음 데이터프레임에서 테마별 PER, PBR의 평균을 계산하세요.​

data = [
     ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
     ["해운", "팬오션", 21.23, 0.95],
     ["시스템반도체", "티엘아이", 35.97, 1.12],
     ["해운", "HMM", 21.52, 3.20],
     ["시스템반도체", "아이에이", 37.32, 3.55],
     ["2차전지(생산)", "LG화학", 83.06, 3.75]
 ]

columns = ["테마", "종목명", "PER", "PBR"]
df9 = pd.DataFrame(data=data, columns=columns)
df9.groupby( by = '테마').mean()


# 10. DataFrame.groupby max, min
# 위 데이터프레임에서 테마별 PER, PBR의 최대, 최소값을 계산하세요.
df10 = pd.DataFrame(data=data,columns= columns)
df10.groupby(by = '테마').max()
df10.groupby(by = '테마').min()

# 11.  DataFrame.groupby size
# 위 데이터프레임에서 테마별 종목의 개수를 계산하세요.
df10.groupby(by = '테마').size()


# 12. DataFrame.groupby get_group
# 위 데이터프레임을 '테마' 컬럼으로 그룹핑한 후 '2차전지(생산)' 그룹만 가져오세요.
gb1 = df10.groupby(by='테마') #테마로 그룹핑하고
gb1.get_group('2차전지(생산)') #get group으로 가져온다


# 13. DataFrame groupby
# 위 데이터프레임에서 테마별 PER의 평균을 계산하세요.

df13 = gb1['PER'].mean() # 테마별 평균
print('13번',df13 )

# 14.  DataFrame.groupby 슬라이싱
# 위 데이터프레임에 대해 테마로 그룹화한 후 PER, PBR 컬럼을 슬라이싱하세요. 그 다음 두 컬럼에 대한 평균을 계산하세요.
df14 = gb1[['PER','PBR']].mean()
print('14번',df14)

# 15. DataFrame.groupby agg
# 위 데이터프레임에 대해서 '테마' 컬럼으로 그룹화한 후 max 함수를 적용하세요.
df10.groupby(by = '테마').agg(max)

# 16. DataFrame.groupby agg
# 위 데이터프레임에 대해서 '테마' 컬럼으로 그룹화한 후 PER 컬럼은 max, PBR 컬럼은 min 함수를 적용해보세요.

df10.groupby(by = '테마').agg({'PER':max , 'PBR':min})

# 17. Group 순회
#위 데이터프레임에 대해 '테마'로 그룹화 한 후 순회하는 다음 코드를 실행하고 결과를 확인하세요. 아래코드 실행해서 결과 보세요.

for name, group in df10.groupby('테마') :
  print(name)
  print( )
  print(group)
# 18. groupby agg
# 위 데이터프레임에 대해서 '테마' 컬럼으로 그룹화한 후 PER, PBR 컬럼에 np.mean과 np.std 함수를 적용하세요.



# 19. nunique
# 위 데이터프레임을 '테마' 컬럼을 기준으로 그룹화한 후 종목명 열만 추출하여 데이터의 빈도수를 계산하세요.
```

* 타이타닉호에서 생존자 , 좌석등급 , 생존여부 등을 알수 있다
컬럼(columns) 설명

survivied: 생존여부 (1: 생존, 0: 사망)

pclass: 좌석 등급 (1등급, 2등급, 3등급)

sex: 성별

age: 나이

sibsp: 형제 + 배우자 수

parch: 부모 + 자녀 수

fare: 좌석 요금

embarked: 탑승 항구 (S, C, Q)

class: pclass와 동일

who: 남자(man), 여자(woman), 아이(child)

adult_male: 성인 남자 여부

deck: 데크 번호 (알파벳 + 숫자 혼용)

embark_town: 탑승 항구 이름

alone: 혼자 탑승 여부

```python
ti.head()
ti['who'].value_counts()

#사용자 함수만들기
#사용자가 만든(정의한)함수를 호출해주기 위해
#.apply(사용자 함수명을 괄호안에 넣어준다)  내용을 알고싶을때 :.value_count()
def transform_who(x):
  if x == 'men':
    return '남자'
  elif x =='women':
    return '여자'
  else :
    return '아이'
ti['who'].apply(transform_who)
ti['who'].apply(transform_who).value_counts()


#좌석요금 : fare
#나이 : age
#구하려는값은 fare / age
def fare_age(y):
  return y['fare'] / y['age']
ti.app = ti.apply(fare_age, axis=1)
ti.head()
```
