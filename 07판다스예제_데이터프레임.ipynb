```python
import pandas as pd
#데이터 프레임 생성하기 방법:1 리스트로
list1 = list([['한빛', '남자', '20', '180'],
       ['한결', '남자', '21', '177'],
       ['한라', '여자', '20', '160']])

col_names=['이름','성명','나이','키']
df1 = pd.DataFrame(list1 , columns= col_names)
df1

df2 = pd.DataFrame(list1 , index= ['1반','2반','3반'])
df2

df2.index
df2.values
print(df2)
print(df2[0]) #0행
print(df2[1]) #1행
print(df2[2]) #2행
print(df2[3]) #3행
```

```python
list1 = [ [ 'KFC',  1000, 4.5],
             [ 'McDonald',  2000, 3.9],
             [ 'SchoolFood',  2500, 4.2],
          ]
col_names = [ 'food', 'price', 'rating' ]
df1 = pd.DataFrame( list1, col_names )
# 딕셔너리를 이용해서 만들기
dic1 = {  'food' : { 0:'KFC', 1: 'McDonald', 2:'SchoolFood' } ,
              'price' : { 0:1000, 1: 2000, 2:2500} ,
              'rating' : { 0:4.5, 1: 3.9, 2:4.2}
          }
df2 = pd.DataFrame( dic1 )
df2
df2[ [  'food' , 'rating'  ]   ]
df2[   'food'  ]   # [   열이름 ]  결과를 시리즈로 보기
df2[   [ 'food']   ]  # [ [ 열이름 ]]  결과를 데이터프레임 보기
df2.rename( columns={ 'food' : 'place'  }, inplace=True)
df2
```
